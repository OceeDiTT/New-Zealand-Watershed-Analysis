### Exercise 2.5: Classifying temperatures

```python
"""
1. Use a for loop and conditional statments (e.g., if, elif, and else) 
to sort the temperatures in the list into the lists associated with each category (see Table 2.7).
2. Answer the following questions:
    2.1 How many times was it cold in Helsinki in April 2013?
    2.2 How many times was it comfortable?
    2.3 Was it ever warm?
"""
temperatures = [-5.4, 1.0, -1.3, -4.8, 3.9, 0.1, -4.4, 4.0, -2.2, -3.9, 4.4,
                -2.5, -4.6, 5.1, 2.1, -2.4, 1.9, -3.3, -4.8, 1.0, -0.8, -2.8,
                -0.1, -4.7, -5.6, 2.6, -2.7, -4.6, 3.4, -0.4, -0.9, 3.1, 2.4,
                1.6, 4.2, 3.5, 2.6, 3.1, 2.2, 1.8, 3.3, 1.6, 1.5, 4.7, 4.0,
                3.6, 4.9, 4.8, 5.3, 5.6, 4.1, 3.7, 7.6, 6.9, 5.1, 6.4, 3.8,
                4.0, 8.6, 4.1, 1.4, 8.9, 3.0, 1.6, 8.5, 4.7, 6.6, 8.1, 4.5,
                4.8, 11.3, 4.7, 5.2, 11.5, 6.2, 2.9, 4.3, 2.8, 2.8, 6.3, 2.6,
                -0.0, 7.3, 3.4, 4.7, 9.3, 6.4, 5.4, 7.6, 5.2]
type(temperatures)

temp_cat= []
for i in temperatures:
    if i < -2:
        temp_cat.append("cold")
    elif -2 <= i < 2:
        temp_cat.append("slippery")
    elif 2 <= i < 15:
        temp_cat.append("comfortable")
    else:
        temp_cat.append("warm")

print(temp_cat)

combined= zip(temperatures, temp_cat)
combined= list(combined)
print(combined)

print(temp_cat.count("cold")) # How many times was it cold in Helsinki in April 2013?: 15 times

print(temp_cat.count("comfortable"))  # How many times was it comfortable?: 59 times

print(temp_cat.count("warm")) # Was it ever warm?: No, it wasnt.
```

### Exercise 2.6: A temperature conversion function

```python
"""
1. Create a function to convert temperature in degrees Fahrenheit to degrees Celsius.
    1.1 Be sure to include a docstring in the function definition.
2. Use the new function to convert the temperatures below to Celsius:
    a. 32 °F
    b. 68 °F
    c. 91 °F
    d. -17 °F
"""

temp= [32, 68, 91, -17]

def Temp_conv_fahr2celsius (temp_fahr):
    
    """
    Convert a temperature from degrees Fahrenheit to degrees Celsius.

    Parameters
    ----------
    fahrenheit : float or int
        Temperature value in degrees Fahrenheit.

    Returns
    -------
    float
        Temperature converted to degrees Celsius.
    """
    temp_celsius= 5 / 9 * (temp_fahr - 32)
    return temp_celsius


print(Temp_conv_fahr2celsius (32))
print(Temp_conv_fahr2celsius (68))    
print(f"{Temp_conv_fahr2celsius (91):.1f}%")    
print(f"{Temp_conv_fahr2celsius (-17):.1f}%")    
```

### Exercise 2.7: A temperature classifier function

```python
"""
1. Create a function that classifies temperatures based on Table 2.7.
    1.1 The function should return the following values for the different categories:
        a. Cold
        b. Slippery
        c. Comfortable
        d. Warm
2. Use the function to calculate the returned value the following temperatures:
    a. 17 °C
    b. 2 °C
    c. 1.9 °C
    d. -2 °C
"""

def Temp_classifier (temp_celsius):
    if temp_celsius < -2:
        print("cold")
    elif -2 <= temp_celsius < 2:
        print("slippery")
    elif 2 <= temp_celsius < 15:
        print("comfortable")
    else:
        print("warm")
    return

Temp_classifier(17)
Temp_classifier(2)
Temp_classifier(1.9)
Temp_classifier(-2)
```
